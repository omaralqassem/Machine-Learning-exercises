```python
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.Age = df.Age.fillna(df.Age.mean())
df
```

```python
df.isna().sum()
```

```python
newdf = df.drop(['Name','Ticket','Cabin','Embarked','PassengerId','SibSp','Parch'],axis='columns')
```

```python
newdf
```

```python
from sklearn.preprocessing import LabelEncoder
Pclass = LabelEncoder()
Sex = LabelEncoder()
Age = LabelEncoder()
Fare = LabelEncoder()
```

```python
newdf['Pclass_n'] = Pclass.fit_transform(newdf['Pclass'])
newdf['Sex_n'] = Sex.fit_transform(newdf['Sex'])
newdf['Age_n'] = Age.fit_transform(newdf['Age'])
newdf['Fare_n'] = Fare.fit_transform(newdf['Fare'])
```

```python
newdf
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.scatter(newdf.Fare_n,newdf.Age_n)
```

```python
target = df['Survived']
target.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(newdf,target,test_size=0.2)
```
