```python
from sklearn import datasets
import pandas as pd
iris = datasets.load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
dataframe = pd.DataFrame(iris.data,columns=iris.feature_names)
```

```python
dataframe
```

```python
dataframe['target'] = iris.target
```

```python
dataframe
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(dataframe.drop('target',axis='columns'),dataframe.target)
```

```python
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.ensemble import RandomForestClassifier
mod = RandomForestClassifier(n_estimators=40)
```

```python
mod.fit(X_train,y_train)
```

```python
mod.score(X_test,y_test)
```

```python
ypred = mod.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cf = confusion_matrix(y_test,ypred)
```

```python
cf
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cf, annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
